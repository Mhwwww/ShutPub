```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

percentage_values = np.linspace(0, 100, 101)
result_values = (20000 * percentage_values / 100 + 100) / 20000


plt.figure(figsize=(10, 6))  # 

sns.set(font_scale=1.2, style='whitegrid', font='CMU Sans Serif')

sns.lineplot(x=(100-percentage_values), y=result_values*100, color='black',label="ShutPub", linewidth = 2)
sns.lineplot(x=percentage_values, y=100, linestyle='--', color='black',label="Baseline", linewidth = 2)

plt.xlabel( "Filtered Message (%)", fontsize=22)
plt.ylabel("Message in the Network (%)", fontsize=22)
plt.title("Network Strain with Filtered Message Percentage", fontsize=24)

plt.legend(loc="upper left")
ax = plt.gca()
ax.set_ylim(0,108)
ax.set_xlim(0,100)


plt.savefig("network_strain.pdf")
```
