```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
import seaborn as sns
import matplotlib as mpl

sns.set(font_scale=1.2, style='whitegrid', font='CMU Sans Serif')


file_set = {
    # #consumer_num: 100 (100 publishers, 1000 ci, 1 filter number)
    # 100:[
    #     '/Users/minghe/Desktop/filterAmount/100.csv',
    #     '/Users/minghe/Desktop/filterAmount/baseline_100.csv',
    # ],
    #
    # #consumer_num: 300 (300 publishers, 1000 ci, 1 filter number)
    # 300:[
    #     '/Users/minghe/Desktop/filterAmount/baseline_300.csv',
    #     '/Users/minghe/Desktop/filterAmount/300.csv',
    # ],

    #consumer_num: 1000 (1000 publishers, 500 ci, 1 filter number)
    1000:[
        'log/filterAmount/1000.csv',
        'log/filterAmount/baseline_1000.csv',
    ],
    1500:[
        'log/filterAmount/1500.csv',
        'log/filterAmount/baseline_1500.csv',
    ],

    2000:[
        'log/filterAmount/2000.csv',
        'log/filterAmount/baseline_2000.csv',
    ],

    2500:[
        'log/filterAmount/2500.csv',
        'log/filterAmount/baseline_2500.csv',
    ],

    3000:[
        'log/filterAmount/3000.csv',
        'log/filterAmount/baseline_3000.csv',
    ],
}

dfs = []

for mi, file_paths in file_set.items():
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        mi_broker = df.copy()
        mi_broker['timestamp'] = mi_broker['Unix Time'] - mi_broker['Unix Time'][0]
        mi_broker['messageInterval'] = mi

        if "baseline" in file_path:
            mi_broker['type'] = "baseline"
        else:
            mi_broker['type'] = "psf"

        dfs.append(mi_broker)

merged_df = pd.concat(dfs)

#
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))

#  markers=["o", "x"]
g = sns.lineplot(data=merged_df, x='messageInterval', y='CPU Utilization (%)', markers=["o", "D"], style='type', errorbar=("ci",95), color="black", style_order=['psf', 'baseline'],hue_order=['psf', 'baseline'], markersize=7, linewidth=1.2)

plt.xlabel('Filter Number', fontsize=22)
plt.ylabel('Broker CPU Utilization (%)', fontsize=22)
plt.title("Broker Strain with Filter Number", fontsize=24)

ax = plt.gca()
ax.set_ylim(0,12)

legend= plt.legend()
legend_labels = legend.texts

for label in legend_labels:
    if label.get_text() == 'psf':
        label.set_text('ShutPub')
    if label.get_text() == 'baseline':
        label.set_text('Baseline')


plt.savefig("filter_amount_lineplot.pdf")
```
