```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

publisher_amount = 1000
msg_per_publisher = 20000
filter_update_per_publisher = 200

percentage_values = np.linspace(0, 100, 101)

# result_values = (20000 * percentage_values / 100 + 1200) / 20000
result_values = (msg_per_publisher * publisher_amount * percentage_values / 100 + publisher_amount + filter_update_per_publisher * publisher_amount) / (msg_per_publisher * publisher_amount)

plt.figure(figsize=(10, 6))  # 

sns.set(font_scale=1.2, style='whitegrid', font='CMU Sans Serif')

sns.lineplot(x=(100-percentage_values), y=result_values*100, color='black',label="ShutPub", linewidth = 2)
sns.lineplot(x=percentage_values, y=100, linestyle='--', color='black',label="Baseline", linewidth = 2)

plt.xlabel( "Filtered Message (%)", fontsize=24)
plt.ylabel("Message in the Network (%)", fontsize=24)
plt.title("Network Strain with Filtered Message Percentage", fontsize=26)

plt.legend(loc="lower left")
# ax = plt.gca()
# ax.set_ylim(0,108)
# ax.set_xlim(0,100)


plt.savefig("network_strain.pdf")
```
